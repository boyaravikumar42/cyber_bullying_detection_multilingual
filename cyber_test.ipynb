```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch

# Path where you saved the model
model_path = "./ML2025/best_model"

# Load model and tokenizer
model = AutoModelForSequenceClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)

# Set model to evaluation mode
model.eval()
```

```python
label_cols = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
```

```python
# Example text for prediction
text = "By pouring water on me and hitting my head.He called me a dumb child.he say bastard.he burn on my hand.he kick on my ass.suck my .. fuck me"

# Tokenize input
inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
inputs
```

```python
with torch.no_grad():  # Disable gradient calculation (inference only)
    outputs = model(**inputs)
    probs = torch.sigmoid(outputs.logits).detach().numpy()
```

```python
print(probs)

print(f"\nText: {text}")
for lbl, score in zip(label_cols, probs[0]):
    print(f"{lbl}: {score:.3f}")
[
```
